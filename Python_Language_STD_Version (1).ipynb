## Arithmatic Operations and Variable Assignment with Python

A variable allows us to store a value or an object which you can use later. When assigning a variable in Python, we usually denote it by = (e.g. `x = 5`; so x gets 5)

Printing in Python using print() and if we want to concatenate, we use a comma.

Arithmatic Operations:
*   Addition: `+`
*   Subtraction: `-`

*   Multiplication: `*`

*   Division: `/`
*   Exponentiation: `**`


*   Modulo: `%`




````python
#Intializing variables
x = 5
y = 2
#Addition +
print(x + y)
#subtraction -
print(x - y)
#division /
print(x/y)
#Multiplication *
print(x * y)
#Modulus %
print(x % y)
#exponential **
print(x ** y)
````

````python
#to print
print(x)
print(y)
#to print with concatenation
print("The value of x is: ", x )
print("The value of y is: ", y )
````

###if else statements 
Notice the following syntax: 
```
if test_expression:
  statement1 
else:
  statement2
```
Notice the tab before statement1 and statement2, this means that we are inside the body of the if and else statements, if that tab is not there, then you are outside















````python
if x >= y: 
  print('x is greater than y')
else:
  print('y is greater than x') 
#########################
if x == y: 
  print('they are equal')
else:
  print('they are not equal') 
#########################
if x != y: 
  print('they are not equal')
else:
  print('they are equal') 
````

### Multiple if…else statements
The  if…else…if statements allow you to execute a block of code among more than 2 alternatives

The syntax of if…else...if statements is:


```
if test_expression1:
  statement1
elif test_expression2: 
  statement2
elif test_expression3: 
  statement3
else:
  statement4
```


Note that we can use "and", "or" to compare more than one condition.
with "and" both conditions must be true to execute. However
with "or", at least one of the conditions must be true to execute

````python
x = 10
if x < 3: 
  print("x is less than 3")
elif x > 3 and x < 5:
  print("x is greater than 3 and less than 5")
elif x > 5 or x == 5:
  print("x is either greater than 5 or equals 5")
else:
  print("x is 3")
````

## Loops

### While loops
An expression is evaluated and the body of the loop is entered if the result is True.
The statements inside the loop are executed and the flow returns to evaluate the expression again.
This is repeated each time until expression evaluates to False, in which case, the loop exits.

The syntax of a while loop: 
```
while test_expression:
  statement
```
Notice the tab before statement, this means that we are inside the body of the loop, if that tab is not there, then you are outside the loop


````python
i = 0
while i < 5:
  print(i)
  i = i +1 
````

````python
a = []
i = 0
while i < 5:
  a.append(i**2)
  i = i +1 
````

````python
a
````

### For Loops
A sequence is a range and i takes on each of its value during the loop. In each iteration, statement is evaluated.

syntax of for loop:
```
for i in sequence:
  statement
```


````python
for i in range(5):
  print(i)
````

````python
for i in range(0,10):
  print(i)
````

````python
for i in range(0,10,2):
  print(i)
````

````python
exl = ['Abdulrahman', 'Ali', 'Omar']
for i in exl:
  print(i)
````

````python
exl = [9, 7, 4, 2, 3, 1]
for i in exl:
  print(i)
````
